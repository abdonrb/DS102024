## Ejercicio 1: Ortogonalidad

Definir una funci칩n en python que detecte si dos vectores son ortogonales

```python
import numpy as np
def son_ortogonales(u,v):
    producto_escalar = np.dot(u, v)
    return producto_escalar == 0
```

```python
v1 = [1, 0]
v2 = [0,1]

son_ortogonales(v1,v2)
```

## Ejercicio 2: Desplazamiento

Definir una funci칩n en python que reciba como par치metro las cuatro esquinas de un cuadril치tero y las desplace 5 unidades hacia arriba y 7 unidades a la izquierda

```python
def desplazar(x1, y1, x2, y2):
  nuevo_x1 = x1 - 7
  nuevo_y1 = y1 + 5
  nuevo_x2 = x2 - 7
  nuevo_y2 = y2 + 5
  return nuevo_x1, nuevo_y1, nuevo_x2, nuevo_y2
```

Ejecutar la siguiente celda descomentando las lineas en verde para verificar su funcionamiento.

```python
import matplotlib.pyplot as plt
x1 = np.array([1,1])
y1 = np.array([1,2])
x2 = np.array([2,3])
y2 = np.array([8,4])

x_coords, y_coords = zip(x1, y1, x2, y2)
plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co")

a_,b_,c_,d_ = desplazar(x1, y1, x2, y2)
x_coords_b, y_coords_b = zip(a_,b_,c_,d_,a_)
plt.plot(x_coords_b, y_coords_b, "b-", x_coords_b, y_coords_b, "bo")

plt.show()
```
